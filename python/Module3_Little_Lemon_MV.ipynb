## Task 1: Connect to the database
In this task, a connection to the Little Lemon MySQL database is established using the MySQL Connector for Python. A cursor object is also created to allow SQL queries to be executed.

```python
# Establish connection
import os
import mysql.connector as connector

connection = connector.connect(
    user=os.getenv("DB_USER"),
    password=os.getenv("DB_PASSWORD"),
    host="localhost",
    database="littlelemondb"
)


cursor = connection.cursor()

print("Connected and cursor created")
```

## Task 2: Display all tables in the database

This task executes a SQL query to display all tables available in the Little Lemon database, verifying that the connection has been successfully established.

```python
show_tables_query = "SHOW TABLES"
cursor.execute(show_tables_query)
```

```python
results = cursor.fetchall()
print(results)
```

## Task 3: Query customer details using a JOIN

In this task, an SQL query is executed using an INNER JOIN to retrieve the full name and contact details of customers who have placed orders with a total cost greater than $60.

```python
query = """
SELECT 
    CONCAT(cd.FirstName, ' ', cd.LastName) AS FullName,
    cd.Phone,
    cd.Email,
    o.TotalCost
FROM CustomerDetails cd
INNER JOIN Orders o
    ON cd.CustomerID = o.CustomerID
WHERE o.TotalCost > 60;
"""

cursor.execute(query)
results = cursor.fetchall()

for row in results:
    print(row)
```
